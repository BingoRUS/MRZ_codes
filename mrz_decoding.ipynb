```
import pytesseract
from google.colab.patches import cv2_imshow
import cv2

pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'
```

```
image = cv2.imread("sample2.jpg")
```

```
cv2_imshow(image)
```

```
mrz_code = pytesseract.image_to_string(image, lang="mrz")[:-2]
```

```
!pip install mrz
```

```
from mrz.checker.td3 import TD3CodeChecker, get_country

td3_check = TD3CodeChecker(mrz_code)

fields = td3_check.fields()

print(fields.name, fields.surname)
print(get_country(fields.country))
print(fields.document_number)
```

```
image = cv2.imread("sample10.jpg")
mrz_code = pytesseract.image_to_string(image, lang="mrz")[:-2]
td3_check = TD3CodeChecker(mrz_code)

fields = td3_check.fields()

print(fields.name, fields.surname)
print(get_country(fields.country))
print(fields.document_number)
```
